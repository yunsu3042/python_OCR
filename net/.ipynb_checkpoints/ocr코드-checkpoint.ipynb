```python
def y_cordinate_(url=None, img=None, arr=None):
    if url:
        arr = image_into_array(url)
    elif img:
        arr = image_into_array_byimg(img)
    elif arr.any():
        arr = arr
    height, width = arr.shape
    linetop = []
    linebot = []
    for h in range(height - 1):
        if 1 in arr[h]:
            linetop.append(max(h-2, 0))
            break
    for h in list(range(1, height)[::-1]):
        if 1 in arr[h]:
            linebot.append(min(h+2, height - 1))
            break
    if len(linebot) > len(linetop):
        linetop.append(0)
    elif len(linetop) > len(linebot):
        linebot.append(height)
    return linetop,linebot

def crop(url):
    ful_img = Image.open(url)
    plt.imshow(ful_img)
    plt.axis("off")
    plt.show()
    lines = y_crop(url=url)
    characters = {}        
    for idx, line in enumerate(lines):
        characters["line" + str(idx + 1)] = []
        width,height = line.size
        cropby, max_wit = x_cordinate(img=line)
        for pos in range(len(cropby[0])):
            char = line.crop((cropby[0][pos], 0, cropby[1][pos], height))
            wit = cropby[1][pos] - cropby[0][pos]
            extra = max_wit - wit 
            arr = np.array(char)
            arr = add_col(arr, where="left", n=extra//2 )
            arr = add_col(arr, where='right', n=extra//2)
            char = Image.fromarray(arr)
            characters["line" + str(idx + 1)].append(char)
    
    return characters

def cross_entropy_error(y, t):
    y = np.exp(y) / np.sum(np.exp(y))
    # y.shape[0]으로 배치사이즈를 가져올 때 1을 가져오기위해서. y.shape (10,) --> (1,10)
    if y.ndim == 1:
        y = y.reshape(1, y.size)
        t = t.reshape(1, t.size)
    output_size = y.shape[1]
    batch_size = y.shape[0]
    h = 1e-7

    # 전제조건 : y는 one_hot_label이다. 따라서 if문의 조건식은 t와 y 모두 one_hot일경우 이다. --> t를 index 라벨로 바꾼다.
    if t.size != y.size:
        temp = np.zeros_like(y)
        for idx, row in enumerate(temp):
            row[t[idx]] = 1
        t = temp
    error = np.sum(-t * np.log(y + h))
    mean_error = error / batch_size
    return mean_error

def crop_(url):
    ful_img = Image.open(url)
    plt.imshow(ful_img)
    plt.axis("off")
    plt.show()
    lines = y_crop(url=url)
    characters = {}
    space_dic = {}
    for idx, line in enumerate(lines):
        characters["line" + str(idx + 1)] = []
        width,height = line.size
        cropby, max_wit = x_cordinate(img=line)
        space_dic['line' + str(idx + 1)] = find_space(cropby)
        for pos in range(len(cropby[0])):
            char = line.crop((cropby[0][pos], 0, cropby[1][pos], height))
            top, bot = y_cordinate_(img=char)
            w, h = char.size
            char = char.crop((0, top[0], w, bot[0]))
            hgt = bot[0] - top[0]
            extra_h = height - hgt
            wit = cropby[1][pos] - cropby[0][pos]
            extra_w = max_wit - wit 
            arr = np.array(char)
            arr = add_row(arr, where='high', n=extra_h//2)
            arr = add_row(arr, where='low', n=extra_h//2)
            arr = add_col(arr, where="left", n=extra_w//2 )
            arr = add_col(arr, where='right', n=extra_w//2)
            char = Image.fromarray(arr)
            characters["line" + str(idx + 1)].append(char)
    
    return characters, space_dic

def find_space(cropby):
    poss = cropby[0].copy(), cropby[1].copy()
    a = poss[0]
    a.append(poss[1][-1])
    b = poss[1]
    b.insert(0,0)
    thre = 5
    space_list = []
    idx = 0
    for x, y in zip(a, b):
        if abs(x - y) > thre:
            space_list.append(idx)
        idx += 1
    return space_list
```

```python
def make_matrix(arr, size=(28, 28)):
    H, W = arr.shape
    sur_H = abs(size[0] - H)
    sur_W = abs(size[0] - W)
    half = sur_H //2
    left = sur_H - half
    if size[0] - H >= 0:
        arr = add_row(arr, where='high', n=half)
        arr = add_row(arr, where='low', n=left)
    elif size[0] - H < 0:
        arr = delete_row(arr, where='high', n=half)
        arr = delete_row(arr, where='low', n=left)
    half = sur_W // 2
    left = sur_W - half
    if size[0] - W >= 0:
        arr = add_col(arr, where='right', n=half)
        arr = add_col(arr, where='left', n=left)
    elif size[0] - H < 0:
        arr = delete_col(arr, where='right', n=half)
        arr = delete_col(arr, where='left', n=left)
    return arr

def make_center(img=None, arr=None):
    if img:
        simarr = image_into_array_byimg(img)
        arr = np.array(img)
    elif arr.any():
        img = Image.fromarray(arr)
        simarr = image_into_array_byimg(img)
    H, W = simarr.shape
    top, bot = y_cordinate_(arr=simarr)
    vac_top = top[0] 
    vac_bot = H - 1 - bot[0]
    plt.imshow(img)
    plt.show()
    print("vac_top: " + str(vac_top))
    print("vac_bot: " + str(vac_bot))
    n = (vac_top - vac_bot)//2
    print(n)
    if n >= 0:
        where=['low', 'high']
    elif n < 0:
        where=['high','low']
    arr = add_row(arr, where=where[0], n=n)
    arr = delete_row(arr, where=where[1], n=n)
    return arr
    
    
def vacant_margin(img):
    img_tmp = img.copy()
    img_tmp.thumbnail((28, 28))
    arr = image_into_array_byimg(img_tmp)
    top, bot = y_cordinate_(arr=arr)
    wit = check_wit(arr=arr)
    heg = bot[0] - top[0]
    H, W = arr.shape
    vw = W - wit
    vh = H - heg
    return vw, vh


def check_wit(img=None, url=None, arr=None):
    if img:
        arr = image_into_array_byimg(img)
    if url:
        arr = image_into_array(url)
    h, w = arr.shape

    for i in range(w - 1):
        if 1 in arr[:, i]:
            st = max(i - 1, 0)
            break
    for i in list(range(1, w))[:: -1]:
        if 1 in arr[:, i]:
            end = min(i + 1, w - 1)
            break
    wit = end - st
    return wit
```

```python
def cross_entropy_error(y,t):
    y = np.exp(y)/np.sum(np.exp(y))
    #y.shape[0]으로 배치사이즈를 가져올 때 1을 가져오기위해서. y.shape (10,) --> (1,10)
    if y.ndim == 1:
        y = y.reshape(1, y.size)
        t = t.reshape(1, t.size)
    output_size = y.shape[1]
    batch_size = y.shape[0]
    h = 1e-7
    
    #전제조건 : y는 one_hot_label이다. 따라서 if문의 조건식은 t와 y 모두 one_hot일경우 이다. --> t를 index 라벨로 바꾼다.
    if t.size != y.size:
        temp = np.zeros_like(y)
        for idx, row in enumerate(temp):
            row[t[idx]] = 1
        t = temp
    error = np.sum(-t * np.log(y + h))
    mean_error = error / batch_size
    return mean_error
```
